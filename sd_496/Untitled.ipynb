```python
import pandas as pd
import os
```

```python
# get all years with 496s, all data is publically available at https://efile.sandiego.gov/help/public/document?doc_id=campaign-bulk-export&doc_type=public
folder_path = "./sd_xlsx/"

# Iterate through all files in the folder
dfs = {}
count = 2016
for filename in os.listdir(folder_path):
    if filename.endswith(".xlsx") or filename.endswith(".xls"):
        file_path = os.path.join(folder_path, filename)
        # Read the Excel file
        df = pd.read_excel(file_path, sheet_name='S496')
        dfs[count] = df
        count += 1
```

```python
# delete all years without and 496's
for i in list(dfs.keys()):
    if dfs[i].shape[0] == 0:
        del dfs[i]        
```

```python
# get all police 496's from every viable year and also reset the index
for i in list(dfs.keys()):
    table = dfs[i]
    dfs[i] = table[table['Filer_NamL'].str.lower().str.contains('police')].reset_index()
```

```python
# drop unneccessary columns
wanted_cols = ['Filer_ID','Filer_NamL', 'Rpt_ID_Num', 'e_filing_id', 'orig_e_filing_id', 'Rpt_Date', 'Amount', 'Exp_Date', 'Expn_Dscr', 'Supp_Opp_Cd', 'Cand_NamL']
for i in list(dfs.keys()):
    dfs[i] = dfs[i][wanted_cols]
```

```python
# export csv
export_path = './export/'
for i in list(dfs.keys()):
    file_path = export_path + str(i) + '_f496.csv'
    dfs[i].to_csv(file_path, index=False)
    
```
